```python
# 1️⃣ 패키지 불러오기
import geopandas as gpd
import matplotlib.pyplot as plt

# 2️⃣ GeoJSON 파일 읽기
burned = gpd.read_file('Burned_Area_2025.geojson')  # 파일 경로 확인! data 폴더에 있다고 가정

# 3️⃣ 데이터 확인
print("Burned Area 데이터 정보:")
print(burned.head())
print("총 폴리곤 개수:", len(burned))

# 4️⃣ 좌표계 확인 (EPSG:4326, WGS84여야 함)
print("좌표계 (CRS):", burned.crs)

# 5️⃣ 시각화
fig, ax = plt.subplots(figsize=(10,10))
burned.plot(ax=ax, color='red', edgecolor='black', alpha=0.5)

plt.title('2025 Burned Area Polygons')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()
```
