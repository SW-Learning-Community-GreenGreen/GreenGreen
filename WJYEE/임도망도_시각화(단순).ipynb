```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Shapefile 파일 경로 (자신의 환경에 맞게 수정)
file_path = "/Users/WJY/Downloads/50/50.shp"

# 파일 읽기
gdf = gpd.read_file(file_path)

# 데이터 확인
print(gdf.head(20))

# 시각화
gdf.plot(edgecolor='black', figsize=(10, 10))
plt.title("임도망도 (임도 Shapefile 시각화)")
plt.axis('off')
plt.show()
```
